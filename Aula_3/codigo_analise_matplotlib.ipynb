```python
df['data']=pd.to_datetime(df['data']) # Converte a coluna data em um objeto datetime (Não transformar fará o computador ficar lento no processamento)
plt.plot(df['data'],df['temperatura'])

plt.title('Temperatura x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Temperatura')
plt.legend('Temperatura',loc='upper left') # Coloca a legenda no grafico no alto, lado esquerdo
#plt.legend('Temperatura',loc='upper right') # Coloca a legenda no grafico no alto, lado esquerdo



```
