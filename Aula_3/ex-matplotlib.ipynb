```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gerando números com numpy
dados_x = np.linspace(0,10,10) # Gera valores com a biblioteca numpy
dados_y = dados_x * dados_x # Gera valores multiplicando um dados pelo outro
```

```python
plt.figure(figsize=(10,5), facecolor='gray')
Ax = plt.axes() # Cria os eixos do gráfico
Ax.set_facecolor('white') # Cria o fundo do gráfico

plt.xlabel('Eixo x',
           fontweight='bold',
           fontsize='large',
           fontfamily='fantasy',
           color='blue',) # Define um título para o eixo x
plt.ylabel('Eixo y',
           fontweight='bold',
           fontsize='large',
           fontfamily='fantasy',
           color='blue',) # Define um título para o eixo y
plt.title('Exemplo Matplotlib') # Define um título para o gráfico

plt.plot(dados_x, dados_y,
         color='red',linewidth=2,
         marker= 'o',linestyle='dashed',markeredgecolor='black'
         )
```
