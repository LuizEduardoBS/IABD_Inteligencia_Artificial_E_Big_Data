```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gerando n√∫meros com numpy
dados_x = np.linspace(0,10,10) # Gera valores com a biblioteca numpy
dados_y = dados_x * dados_x # Gera valores multiplicando um dados pelo outro
```

```python
plt.figure(figsize=(10,5), facecolor='gray')
Ax = plt.axes() # Cria os eixos do gr√°fico
Ax.set_facecolor('white') # Cria o fundo do gr√°fico

plt.xlabel('Eixo x',
           fontweight='bold',
           fontsize='large',
           fontfamily='fantasy',
           color='blue',) # Define um t√≠tulo para o eixo x
plt.ylabel('Eixo y',
           fontweight='bold',
           fontsize='large',
           fontfamily='fantasy',
           color='blue',) # Define um t√≠tulo para o eixo y
plt.title('Exemplo Matplotlib') # Define um t√≠tulo para o gr√°fico

plt.plot(dados_x, dados_y,
         color='red',linewidth=2,
         marker= 'o',linestyle='dashed',markeredgecolor='black'
         )
```

```python
# Grafico !
plt.rcParams.update({'font.size':15}) # Define o tamanho da fonte
plt.figure(figsize=(12,12)) #Define o tamanho da figura
linhas=2 # Define o n√∫mero de linhas
colunas=2 # Define o n√∫mero de colunas
plt.subplot(linhas,colunas,1) # Define o n√∫mero de linhas e colunas
plt.plot(dados_x,'r--')
plt.title('Grafico 1')

# Grafico 2
plt.subplot(linhas,colunas,2)
plt.plot(dados_x*dados_x,'b-.')
plt.title('Grafico 2') 

# Grafico 3
plt.subplot(linhas,colunas,3)
plt.plot(dados_x/2,'c-.')
plt.title('Grafico 3')

# Grafico 4
plt.subplot(linhas,colunas,4)
plt.plot(dados_y*2,'r--')
plt.title('Grafico 4')

# Comando para salvar a figura
plt.savefig('Exemplo de subplot.jpg',dpi=300,transparent=True, bbox_inches='tight')
```

```python
# Gr√°fico de barras
Figura = plt.figure() # Armazena na variavel Figura a figura a ser plotada
Grafico = Figura.add_axes([0,0,0.7,0.7]) # Cria a grade do gr√°fico
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] # Cria os r√≥tulos
Grafico.bar(Rotulos,dados_x)
plt.title('Quantidade x categoria')
plt.xlabel('Itens')
plt.ylabel('Quantidade')
```

```python
Figura = plt.figure() # Armazena na variavel Figura a figura a ser plotada
Grafico = Figura.add_axes([0,0,0.7,0.7]) # Cria a grade do gr√°fico
Rotulos = ['A','B','C','D','E','F','G','H','I','J'] # Cria os r√≥tulos
Grafico.bar(Rotulos,dados_x,color='blue')
Grafico.bar(Rotulos,dados_x*0.1,color='red')
plt.title('Quantidade x categoria')
plt.xlabel('Itens')
plt.ylabel('Quantidade')
```
