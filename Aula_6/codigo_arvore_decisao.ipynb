```python
import numpy as np # Numpy para cálculos matemáticos
# pip install scikit-learn
from sklearn.model_selection  import train_test_split # Biblioteca para dividior os dados teste e treino
from sklearn.tree import DecisionTreeClassifier # Importa o algoritmo da arvore de decisão
from sklearn import tree # Biblioteca para plotar a arvore de decisão
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score # Métrica de acurácia do algoritmo (acurácia= acertos / acertos+erros)
from sklearn.metrics import precision_score # Métrica de precisão do algoritmo
from sklearn.metrics import recall_score # Métrica de recall do algoritmo
from sklearn.metrics import f1_score # Métrica de f1_score
from sklearn.metrics import confusion_matrix # Importa a matriz de confusão

# pip install scikit-learn
```

```python
# Criando o conjunto de dados
notas = np.array([8,6,5,9,4,3,7,2,1,5,6,9,8])
classes = np.array(['bom','ruim','ruim','bom','pessimo',
                    'pessimo','bom','pessimo',
                     'pessimo','ruim','ruim','bom','bom'])
```

```python
# Dividir os dados em treino e teste
notas_treino, notas_teste, classes_treino, classes_teste = train_test_split(notas.reshape(-1,1),
                                                                            classes.reshape(-1,1), 
                                                                            test_size=0.2, 
                                                                            random_state=42)
```

```python
modelo = DecisionTreeClassifier() # Cria um modelo de decisao
modelo.fit(notas_treino,classes_treino) # Treina o modelo da arvore de decisao
```

```python
previsoes = modelo.predict(notas_teste) # Dentro do modelo predict é colocado as notas de teste
previsoes
```

```python
notas_teste
```

```python
nota = np.array([4]).reshape(-1,1)
resultado = modelo.predict(nota)
resultado
```

```python
fig = plt.figure(figsize=(10,8)) # Cria uma figura de tamnho 10 x 8
# tree.plot_tree
tree.plot_tree(modelo, feature_names=notas.tolist(), class_names=classes.tolist(), filled=True)
plt.title('Arvore de decisão para classificar uma bebida \n'
          'de acordo com a nota', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.grid()
plt.title('Box-plot das notas e classes das bebidas')

sns.boxplot(x = classes, y = notas)
```

```python
# Grafico de barras
plt.figure(figsize=(10,8))
sns.barplot(y=notas, x=classes)
plt.title('Notas de bebidas', fontsize=20)
plt.xlabel('Classificação', fontsize=16)
plt.ylabel('Notas', fontsize=16)
```

```python
# Métrica do sistema
acuracy = accuracy_score(classes_teste,previsoes)
acuracy
```

```python
precision = precision_score(classes_teste, previsoes, pos_label='bom')
precision
```

```python
recall = recall_score(classes_teste, previsoes, pos_label='ruim')
recall
```

```python
f1score = f1_score(classes_teste, previsoes, pos_label='bom')
f1score
```
