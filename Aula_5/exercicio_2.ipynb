```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
Q = np.array([100, 200, 300, 400, 500, 600, 700]).reshape(-1, 1)
T = np.array([30, 35, 40, 45, 60, 100, 150])

# Cria o modelo
modelo = LinearRegression()
modelo.fit(Q, T)
```

```python
# Dados teste
qtde_calor_teste = np.array([150, 450, 850, 1100]).reshape(-1, 1)

temperatura_prevista = modelo.predict(qtde_calor_teste)
```

```python
#Imprimir as previsoes
for i in range(len(qtde_calor_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Qtde calor: {qtde_calor_teste[i][0]} - Temperatura prevista:{temperatura_prevista[i]}")
    #gráfico de pontos -scatter
    plt.scatter(Q, T, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
    plt.plot(qtde_calor_teste, temperatura_prevista, color = 'red', label='Dados de teste')
    plt.xlabel('Temperatura')
    plt.ylabel('Qtde de calor')
    plt.title('Relação entre qtde de calor e temperatura')
    plt.legend()
    plt.grid(True)
    plt.show
```
