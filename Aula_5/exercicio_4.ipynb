```python
import numpy as np
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)
gastos = np.array([750, 1500, 4000, 6000, 18000])

modelo = LinearRegression()
modelo.fit(renda, gastos)
```

```python
# Dados de teste
renda_teste = np.array([3000, 15000, 25000]).reshape(-1, 1)

# Cria as previsões
gasto_previsto = modelo.predict(renda_teste)
```

```python
#Imprimir as previsoes
for i in range(len(renda_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Renda: R${renda_teste[i][0]} - Gasto previsto: R${gasto_previsto[i]}")
    #gráfico de pontos -scatter
    plt.scatter(renda,gastos, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
    plt.plot(renda_teste,gasto_previsto, color = 'red', label='Dados de teste')
    plt.xlabel('Renda')
    plt.ylabel('Gastos')
    plt.title('Relação entre renda e gastos')
    plt.legend()
    plt.grid(True)
    plt.show
```
