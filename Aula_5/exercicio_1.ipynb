```python
import numpy as np
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
vazao = np.array([10, 20, 30,5, 35, 40]).reshape(-1, 1) # Vazão de água em m3/min
nivel = np.array([50, 60, 70, 45, 80, 85]) # Nível do tanque em metros

# Cria e treina modelo de regressão linear
modelo = LinearRegression()
modelo.fit(vazao, nivel) # Cria um modelo com base nos dados de entrada
```

```python
# Dados de teste
vazao_teste = np.array([15,25,33]).reshape(-1, 1) #Vazao da água em m³/min
#Fazer previsões com base na vazao de água
nivel_previsto = modelo.predict(vazao_teste)
```

```python
#Imprimir as previsoes
for i in range(len(vazao_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Vazão: {vazao_teste[i][0]}m³/min - Nível previsto:{nivel_previsto[i]}")
    #gráfico de pontos -scatter
    plt.scatter(vazao,nivel, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
    plt.plot(vazao_teste,nivel_previsto, color = 'red', label='Dados de teste')
    plt.xlabel('Vazão de Água(m3/min)')
    plt.ylabel('Nível do tanque (metros)')
    plt.title('Relação entre vazão de água e nível do tanque')
    plt.legend()
    plt.grid(True)
    plt.show
```
