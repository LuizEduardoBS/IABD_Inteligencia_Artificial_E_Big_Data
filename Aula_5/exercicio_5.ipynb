```python
import numpy as np
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1) 
demanda = np.array([100, 200, 80, 100, 160]) 

modelo = LinearRegression()
modelo.fit(consumo, demanda)
```

```python
# Dados de teste
consumo_teste = np.array([15, 90, 110, 250, 300]).reshape(-1, 1)

# Cria a previsão
demanda_prevista = modelo.predict(consumo_teste)
```

```python
#Imprimir as previsoes
for i in range(len(consumo_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Consumo: {consumo_teste[i][0]}kw - Demanda prevista:{demanda_prevista[i]}")
    #gráfico de pontos -scatter
    plt.scatter(consumo,demanda, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
    plt.plot(consumo_teste,demanda_prevista, color = 'red', label='Dados de teste')
    plt.xlabel('Consumo (kw)')
    plt.ylabel('Demanda')
    plt.title('Relação entre consumo de energia e demanda')
    plt.legend()
    plt.grid(True)
    plt.show
```
