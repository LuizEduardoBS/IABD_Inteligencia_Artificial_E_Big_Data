```python
import numpy as np
from sklearn.linear_model import LinearRegression # da biblioteca sklearn importa o modelo de regressao linear
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
area = np.array([50, 100, 120, 300]).reshape(-1, 1)
valor = np.array([180000, 300000, 375000, 600000])

# Cria e treina o modelo
modelo = LinearRegression()
modelo.fit(area, valor)
```

```python
# Dados de teste
area_teste = np.array([75, 180, 420, 500]).reshape(-1, 1)
# Faz as previsões
area_prevista = modelo.predict(area_teste)
```

```python
#Imprimir as previsoes
for i in range(len(area_teste)): #Laço de repetição para exibir os dados de saida
    print(f"Área: {area_teste[i][0]}m2 - Valor previsto: R${area_prevista[i]}")
    #gráfico de pontos -scatter
    plt.scatter(area,valor, color='blue' , label='Dados de Treinamento')
    #gráfico de linhas
    plt.plot(area_teste,area_prevista, color = 'red', label='Dados de teste')
    plt.xlabel('Área (m2)')
    plt.ylabel('Valor R$')
    plt.title('Relação entre área e valor do imovel')
    plt.legend()
    plt.grid(True)
    plt.show
```
