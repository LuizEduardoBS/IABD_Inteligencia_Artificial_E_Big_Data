```python
import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Player/Desktop/clone/IABD_Inteligencia_Artificial_E_Big_Data/Aula_2_Matplotlib/Base de dados/monitoramento_tempo.csv")
```

```python
# EXERCÍCIO 1

df['data']=pd.to_datetime(df['data'])
plt.figure(figsize=(10,6))
plt.plot(df['data'],df['humidade'])
plt.title('Humidade x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Humidade')
plt.legend('Humidade',loc='lower right')

# Comentario pedido no exercício 3
plt.text(df['data'].iloc[0], df['humidade'].iloc[0], 'Este ponto é importante',
         fontsize=12, color='red', horizontalalignment='left', verticalalignment='bottom')

```

```python
# EXERCÍCIO 2

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['velocidade do vento'])
plt.title('Velocidade do vento x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Velocidade do vento')
plt.legend('Velocidade do vento', loc='upper right')
```

```python
# EXERCÍCIO 3

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['humidade'])
plt.title('Humidade x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Humidade')
plt.legend(['Humidade'], loc='lower right')

# Anotação no gráfico
plt.text(df['data'].iloc[0], df['humidade'].iloc[0] - 1, 'Início do período', 
         fontsize=12, color='red', verticalalignment='top')
```

```python
# EXERCÍCIO 4

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['direção do vento'])
plt.title('Direção do Vento x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Direção do Vento')
plt.legend(['Direção do Vento'], loc='upper right')

# Anotação no gráfico
plt.text(df['data'].iloc[10], df['direção do vento'].iloc[10] + 5, 'Mudança significativa', 
         fontsize=12, color='blue', verticalalignment='bottom')
```

```python
# EXERCÍCIO 5

plt.figure(figsize=(10,6))
plt.plot(df['data'], df['pressão'])
plt.title('Pressão x Ano', fontsize=15)
plt.xlabel('Ano')
plt.ylabel('Pressão')
plt.legend(['Pressão'], loc='upper right')

# Anotação no gráfico
plt.text(df['data'].iloc[5], df['pressão'].iloc[5] + 1, 'Aumento de pressão', 
         fontsize=12, color='red', verticalalignment='bottom')
```

```python
# EXERCÍCIO 9

df_periodo1 = df[(df['data'] >= '2016-05-01') & (df['data'] <= '2016-08-31')]
df_periodo2 = df[(df['data'] >= '2016-01-01') & (df['data'] <= '2016-12-31')]

plt.figure(figsize=(10,6))
plt.subplot(2, 1, 1)
plt.plot(df_periodo1['data'], df_periodo1['temperatura'])
plt.title('Temperatura - Período 05-2016 a 08-2016', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Temperatura')

plt.subplot(2, 1, 2)
plt.plot(df_periodo2['data'], df_periodo2['temperatura'])
plt.title('Temperatura - Ano 2016', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Temperatura')

plt.tight_layout() # é utilizado no Matplotlib para ajustar automaticamente os parâmetros de subplots de forma que o layout seja organizado e sem sobreposição de elementos.
```

```python
# EXERCÍCIO 10

df_periodo = df[(df['data'] >= '2017-05-01') & (df['data'] <= '2017-08-31')]

plt.figure(figsize=(10,6))

# Gráfico maior - Temperatura
plt.subplot(2, 1, 1)
plt.plot(df_periodo['data'], df_periodo['temperatura'])
plt.title('Temperatura - Período 05-2017 a 08-2017', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Temperatura')

# Gráfico menor - Umidade
plt.subplot(2, 1, 2)
plt.plot(df_periodo['data'], df_periodo['humidade'])
plt.title('Umidade - Período 05-2017 a 08-2017', fontsize=15)
plt.xlabel('Data')
plt.ylabel('Umidade')

plt.tight_layout() # é utilizado no Matplotlib para ajustar automaticamente os parâmetros de subplots de forma que o layout seja organizado e sem sobreposição de elementos.
```
